## Informe Final de Análisis de Tiendas

### Introducción

El objetivo de este análisis es evaluar el rendimiento de las cuatro tiendas de Alura Store para identificar cuál de ellas presenta menor eficiencia y debería ser vendida por el Sr. Juan para iniciar un nuevo emprendimiento.

### Análisis de datos

Se analizaron los siguientes factores:

- Ingresos totales por tienda
- Categorías de productos más vendidas
- Calificación promedio de los clientes
- Productos más y menos vendidos
- Costo promedio de envío

Los gráficos generados permitieron visualizar el rendimiento de cada tienda y comparar sus resultados.

### Resultados

Los ingresos muestran diferencias entre las tiendas, evidenciando cuál genera menor facturación.

Las categorías de productos revelan qué tipo de artículos tienen mayor demanda en cada tienda.

Las calificaciones promedio de los clientes reflejan el nivel de satisfacción de los compradores.

Además, se identificaron los productos más vendidos y los menos vendidos, permitiendo evaluar el comportamiento del mercado.

Finalmente, el análisis del costo promedio de envío permitió entender el impacto logístico en cada tienda.

### Conclusión

Con base en el análisis realizado, la tienda con menor rendimiento general debería ser vendida. Esta decisión se basa en la combinación de menores ingresos, menor volumen de ventas y menor satisfacción del cliente en comparación con las otras tiendas.

Por lo tanto, se recomienda vender la tienda que presentó los resultados más bajos en estos indicadores para optimizar la operación del negocio y enfocar los recursos en las tiendas más rentables.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_tienda1 = tienda["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingresos Tienda 1:", ingresos_tienda1)
print("Ingresos Tienda 2:", ingresos_tienda2)
print("Ingresos Tienda 3:", ingresos_tienda3)
print("Ingresos Tienda 4:", ingresos_tienda4)
```

# 2. Ventas por categoría

```python
categoria_t1 = tienda["Categoría del Producto"].value_counts()
categoria_t2 = tienda2["Categoría del Producto"].value_counts()
categoria_t3 = tienda3["Categoría del Producto"].value_counts()
categoria_t4 = tienda4["Categoría del Producto"].value_counts()
```

```python
print("Categorías Tienda 1")
print(categoria_t1)

print("\nCategorías Tienda 2")
print(categoria_t2)

print("\nCategorías Tienda 3")
print(categoria_t3)

print("\nCategorías Tienda 4")
print(categoria_t4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_t1 = tienda["Calificación"].mean()
calificacion_t2 = tienda2["Calificación"].mean()
calificacion_t3 = tienda3["Calificación"].mean()
calificacion_t4 = tienda4["Calificación"].mean()
```

```python
print("Calificación promedio Tienda 1:", calificacion_t1)
print("Calificación promedio Tienda 2:", calificacion_t2)
print("Calificación promedio Tienda 3:", calificacion_t3)
print("Calificación promedio Tienda 4:", calificacion_t4)
```

# 4. Productos más y menos vendidos

```python
productos_t1 = tienda["Producto"].value_counts()
productos_t2 = tienda2["Producto"].value_counts()
productos_t3 = tienda3["Producto"].value_counts()
productos_t4 = tienda4["Producto"].value_counts()
```

```python
print("Tienda 1")
print("Más vendido:", productos_t1.idxmax())
print("Menos vendido:", productos_t1.idxmin())

print("\nTienda 2")
print("Más vendido:", productos_t2.idxmax())
print("Menos vendido:", productos_t2.idxmin())

print("\nTienda 3")
print("Más vendido:", productos_t3.idxmax())
print("Menos vendido:", productos_t3.idxmin())

print("\nTienda 4")
print("Más vendido:", productos_t4.idxmax())
print("Menos vendido:", productos_t4.idxmin())
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()
```

```python
print("Costo promedio envío Tienda 1:", envio_t1)
print("Costo promedio envío Tienda 2:", envio_t2)
print("Costo promedio envío Tienda 3:", envio_t3)
print("Costo promedio envío Tienda 4:", envio_t4)
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]

plt.bar(tiendas, ingresos)

plt.title("Ingresos por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

```python
calificaciones = [calificacion_t1, calificacion_t2, calificacion_t3, calificacion_t4]

plt.bar(tiendas, calificaciones)

plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación")

plt.show()
```
